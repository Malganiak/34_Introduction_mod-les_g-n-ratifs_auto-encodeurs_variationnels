* A quoi sert principalement un auto-encodeur variationnel ?

Un auto-encodeur variationnel est un outil polyvalent utilisé pour diverses tâches, allant de la compression de données à la génération de nouvelles données en passant par l'apprentissage semi-supervisé. Son utilité réside dans sa capacité à apprendre une représentation efficace et probabiliste des données, ce qui lui permet d'être appliqué à un large éventail de problèmes d'apprentissage automatique.

* Décrire brièvement les notions d'espace latent et de vecteur latent. A quoi correspondent-ils dans le cas de l'auto-encodeur variationnel ?


L'espace latent:
dans le contexte des auto-encodeurs, y compris les auto-encodeurs variationnels (VAE), fait référence à une représentation de données abstraite et de dimensions réduites. C'est un espace dans lequel les données sont encodées après avoir traversé la couche d'encodeur du réseau neuronal. Cette représentation est souvent caractérisée par des variables latentes qui captent les caractéristiques essentielles des données, mais qui peuvent ne pas être directement observables dans les données d'entrée..

Le vecteur lat :
nt est simplement une expression plus concrète de cette représentation dans l'espace latent. C'est un vecteur numérique qui représente une observation ou une instance de données dans cet espace latent. Chaque composante de ce vecteur correspond à une dimension de l'espace latent et encode une caractéristique particulière de la donnée d'entrée. En d'autres termes, chaque vecteur latent représente une version condensée et abstraite des données originales dans un espace de dimension réduite. Ce vecteur latent est ensuite utilisé par la couche de décodeur pour reconstruire une version de l'entrée initiale, souvent avec pour objectif de minimiser la perte de reconstruction.

* Qu'est-ce que et surtout à quoi sert la "Kullback-Leibler (KL) divergence" ? (vulgariser)

La divergence de Kullback-Leibler (KL) est une mesure qui nous dit à quel point deux distributions de probabilité sont différentes l'une de l'autre.

Imaginez deux distributions de probabilité comme des manières différentes de voir le monde. La divergence de KL nous dit à quel point ces deux visions sont éloignées l'une de l'autre.

Pourquoi est-ce important ? Eh bien, dans le domaine de l'apprentissage automatique, en particulier dans les auto-encodeurs variationnels (VAE), nous voulons souvent que notre modèle génère des données qui ressemblent le plus possible à celles que nous lui avons données à l'entraînement.

La KL-divergence est utilisée dans les VAE pour nous aider à obtenir cela. Plus précisément, elle nous aide à contraindre l'espace latent, où les données sont représentées de manière compressée, à avoir une certaine forme. Cela signifie que le VAE est encouragé à générer des données qui "collent" à une certaine structure ou à une certaine distribution, ce qui peut être très utile dans la génération de données réalistes. En d'autres termes, la KL-divergence est un outil important pour s'assurer que les données générées par un modèle sont cohérentes et de bonne qu

## Construction de l'autoencodeur variationnel (dimension de l'espace latent = 2)

```python
class CVAE(tf.keras.Model):
  """Convolutional variational autoencoder."""

  def __init__(self, latent_dim):
    super(CVAE, self).__init__()
    self.latent_dim = latent_dim
    
    # Définition de l'encodeur
    self.encoder = tf.keras.Sequential(
        [
            tf.keras.layers.InputLayer(input_shape=(28, 28, 1)),  # Couche d'entrée pour les images
            tf.keras.layers.Conv2D(
                filters=32, kernel_size=3, strides=(2, 2), activation='relu'),  # Première couche de convolution
            tf.keras.layers.Conv2D(
                filters=64, kernel_size=3, strides=(2, 2), activation='relu'),  # Seconde couche de convolution
            tf.keras.layers.Flatten(),  # Aplatir les données pour les connecter à une couche Dense
            # Pas d'activation finale, les sorties sont les paramètres de la distribution latente
            tf.keras.layers.Dense(latent_dim + latent_dim),  # Couche dense pour générer les paramètres de la distribution latente
        ]
    )

    # Définition du décodeur
    self.decoder = tf.keras.Sequential(
        [
            tf.keras.layers.InputLayer(input_shape=(latent_dim,)),  # Couche d'entrée pour les échantillons latents
            tf.keras.layers.Dense(units=7*7*32, activation=tf.nn.relu),  # Couche dense pour transformer les échantillons en une forme compatible avec la convolution
            tf.keras.layers.Reshape(target_shape=(7, 7, 32)),  # Remodeler les données pour les convolutions transposées
            tf.keras.layers.Conv2DTranspose(
                filters=64, kernel_size=3, strides=2, padding='same',
                activation='relu'),  # Première couche de convolution transposée
            tf.keras.layers.Conv2DTranspose(
                filters=32, kernel_size=3, strides=2, padding='same',
                activation='relu'),  # Seconde couche de convolution transposée
            # Pas d'activation finale, les sorties sont les logits pour la reconstruction
            tf.keras.layers.Conv2DTranspose(
                filters=1, kernel_size=3, strides=1, padding='same'),  # Couche de sortie pour reconstruire l'image
        ]
    )

  @tf.function
  def sample(self, eps=None):
    """Fonction pour échantillonner des échantillons latents."""
    if eps is None:
      eps = tf.random.normal(shape=(100, self.latent_dim))  # Échantillonnage de bruit aléatoire
    return self.decode(eps, apply_sigmoid=True)  # Décodage des échantillons latents en images

  def encode(self, x):
    """Fonction pour encoder des images en paramètres de distribution latente."""
    mean, logvar = tf.split(self.encoder(x), num_or_size_splits=2, axis=1)  # Division des sorties en moyenne et variance logarithmique
    return mean, logvar  # Renvoie la moyenne et la variance logarithmique

  def reparameterize(self, mean, logvar):
    """Fonction pour rééchantillonner en utilisant les paramètres de la distribution latente."""
    eps = tf.random.normal(shape=mean.shape)  # Échantillonnage de bruit aléatoire
    return eps * tf.exp(logvar * .5) + mean  # Rééchantillonnage en utilisant la formule de réparamétrisation

  def decode(self, z, apply_sigmoid=False):
    """Fonction pour décoder les échantillons latents en images."""
    logits = self.decoder(z)  # Passage des échantillons latents à travers le décodeur
    if apply_sigmoid:
      probs = tf.sigmoid(logits)  # Application de la fonction sigmoid si nécessaire
      return probs  # Renvoie les probabilités si demandé
    return logits  # Renvoie les logits, utilisés pour la reconstruction
```

## Définition de la fonction de perte et l'optimiseur

```python
# Définition de l'optimiseur Adam avec un taux d'apprentissage de 1e-4
optimizer = tf.keras.optimizers.Adam(1e-4)

# Fonction pour calculer la log-probabilité d'un échantillon selon une distribution normale
def log_normal_pdf(sample, mean, logvar, raxis=1):
  log2pi = tf.math.log(2. * np.pi)
  return tf.reduce_sum(
      -.5 * ((sample - mean) ** 2. * tf.exp(-logvar) + logvar + log2pi),
      axis=raxis)

# Fonction pour calculer la perte du modèle
def compute_loss(model, x):
  # Encoder l'image d'entrée x pour obtenir la moyenne et le log de la variance de la distribution latente
  mean, logvar = model.encode(x)
  # Rééchantillonnage de la distribution latente
  z = model.reparameterize(mean, logvar)
  # Décodage de z pour obtenir la sortie reconstruite
  x_logit = model.decode(z)
  # Calcul de l'entropie croisée entre la sortie reconstruite et l'image d'entrée originale
  cross_ent = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_logit, labels=x)
  # Calcul de la log-probabilité conditionnelle de x selon z
  logpx_z = -tf.reduce_sum(cross_ent, axis=[1, 2, 3])
  # Calcul de la log-probabilité prior de z
  logpz = log_normal_pdf(z, 0., 0.)
  # Calcul de la log-probabilité postérieure de z
  logqz_x = log_normal_pdf(z, mean, logvar)
  # Retourner la perte totale, qui est la somme des différentes composantes
  return -tf.reduce_mean(logpx_z + logpz - logqz_x)

# Définition d'une fonction pour exécuter une étape d'entraînement
@tf.function
def train_step(model, x, optimizer):
  """Exécute une étape d'entraînement et retourne la perte.

  Cette fonction calcule la perte et les gradients, et utilise ces derniers pour
  mettre à jour les paramètres du modèle.
  """
  with tf.GradientTape() as tape:
    # Calcul de la perte
    loss = compute_loss(model, x)
  # Calcul des gradients
  gradients = tape.gradient(loss, model.trainable_variables)
  # Mise à jour des paramètres du modèle en utilisant l'optimiseur
  optimizer.apply_gradients(zip(gradients, model.trainable_variables))
```

## Entrainement du modèle

```python
epochs = 10  # Nombre d'époques d'entraînement
latent_dim = 2  # Dimension de l'espace latent
num_examples_to_generate = 16  # Nombre d'exemples à générer

random_vector_for_generation = tf.random.normal(
    shape=[num_examples_to_generate, latent_dim])  # Vecteur aléatoire pour la génération d'images
model = CVAE(latent_dim)  # Initialisation du modèle CVAE
```

## Génération d'images synthétiques (utilisation du modèle en inférence)

```python
# Fonction pour générer et sauvegarder des images synthétiques
def generate_and_save_images(model, epoch, test_sample):
    mean, logvar = model.encode(test_sample)  # Encodage de l'échantillon de test
    z = model.reparameterize(mean, logvar)    # Rééchantillonnage dans l'espace latent
    predictions = model.sample(z)             # Décodage des échantillons dans l'espace d'observation

    fig = plt.figure(figsize=(4, 4))  # Création d'une figure

    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i + 1)
        plt.imshow(predictions[i, :, :, 0], cmap='gray')  # Affichage de chaque image
        plt.axis('off')  # Désactivation des axes

    plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))  # Sauvegarde de la figure
    plt.show()  # Affichage de la figure
```

```python
# Sélection d'un échantillon du jeu de test pour la génération d'images
assert batch_size >= num_examples_to_generate
for test_batch in test_dataset.take(1):
    test_sample = test_batch[0:num_examples_to_generate, :, :, :]

generate_and_save_images(model, 0, test_sample)  # Génération et sauvegarde des images initiales
```

```python
from tqdm import tqdm

# Boucle d'entraînement sur plusieurs époques
for epoch in tqdm(range(1, epochs + 1), desc='Training progress', unit='epoch'):
    start_time = time.time()  # Mesure du temps de début de l'époque
    for train_x in train_dataset:  # Itération sur les batches du jeu de données d'entraînement
        train_step(model, train_x, optimizer)  # Entraînement du modèle sur un batch
    end_time = time.time()  # Mesure du temps de fin de l'époque

    loss = tf.keras.metrics.Mean()  # Initialisation d'une métrique pour calculer la perte
    for test_x in test_dataset:  # Itération sur les batches du jeu de données de test
        loss(compute_loss(model, test_x))  # Calcul de la perte sur un batch
    elbo = -loss.result()  # Calcul de l'ELBO sur l'ensemble de test
    display.clear_output(wait=False)  # Effacement de l'affichage précédent
    print('Epoch: {}, Test set ELBO: {}, time elapse for current epoch: {}'
          .format(epoch, elbo, end_time - start_time))  # Affichage de l'ELBO et du temps écoulé
    generate_and_save_images(model, epoch, test_sample)  # Génération et sauvegarde des images générées
```
